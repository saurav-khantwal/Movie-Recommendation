```python
def convert(x):
    temp = []
    for i in ast.literal_eval(x):
        temp.append(i['name'])
    return temp
```

```python
# Now we will process the cast column
def convert_cast(x):
    count = 0
    temp = []
    for i in ast.literal_eval(x):
        if count > 2:
            break
        temp.append(i['name'])
        count += 1
    return temp
```

```python
final_data['cast'] = final_data['cast'].apply(convert_cast)
```

```python
final_data.iloc[0]
```

```python
#Now we will apply preprocessing for the director
ast.literal_eval(final_data.crew.iloc[0])
```

```python
def convert_director(x):
    temp = []
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            temp.append(i['name'])
            return temp
    
```

```python
final_data['director'] = final_data.crew.apply(convert_director)
```

```python
#We can see their are some words which will have same meaning 
#example actor, actors, team, teams
#We will make them into a single word using steming technique

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    x = []
    for i in text.split():
        x.append(ps.stem(i))
    return " ".join(x)
```

```python
new_data['tags'] = new_data.tags.apply(stem)
new_data.head()
```

```python
#Removing stop words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
vectors = cv.fit_transform(new_data['tags'])
```

```python
vectors.shape
```

```python
# vectors[0]
```

```python
## Now to check the similarity between the vectors we will calculate the cosine distance because it is better in case
## Of higher dimensionality data
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index = new_data[new_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    sorted_movies = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:11]
    movie_list = []
    for i in sorted_movies:
        movie_list.append(new_data['title'][i[0]])
    return movie_list
```
